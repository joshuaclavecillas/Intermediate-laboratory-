```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, chi2

sabastian = np.array([0.12,0.15,0.13,0.135,0.08,0.155,0.135,0.165,0.185,0.15,
                      0.165,0.15,0.14,0.175,0.14,0.12,0.11,0.115,0.15,0.155])

joshua = np.array([0.15,0.185,0.155,0.135,0.15,0.17,0.095,0.115,0.185,0.165,
                   0.18,0.15,0.15,0.14,0.17,0.16,0.155,0.17,0.16,0.16])

data = joshua
name = "Joshua"

count = len(data)
mean = np.sum(data)/count

differences = []
for x in data:
    difference = x - mean
    differences.append(difference**2)

summation = np.sum(differences)
standard_deviation = (summation/(count-1))**(1/2)

print(f"{name} mean = {mean}")
print(f"{name} standard deviation = {standard_deviation}")

bin_width = standard_deviation/2
minimum = np.min(data)
maximum = np.max(data)

bins = np.arange(minimum, maximum + bin_width, bin_width)

observed, bin_edges = np.histogram(data, bins=bins)

expected = []
for i in range(len(bin_edges)-1):
    lower = bin_edges[i]
    upper = bin_edges[i+1]

    probability = norm.cdf(upper, mean, standard_deviation) - norm.cdf(lower, mean, standard_deviation)
    expected_value = count * probability

    expected.append(expected_value)

expected = np.array(expected)

print("\nBin analysis")
for i in range(len(observed)):
    print(f"{bin_edges[i]:.4f} to {bin_edges[i+1]:.4f}: observed = {observed[i]}, expected = {expected[i]}")

chi_square_terms = []
for o, e in zip(observed, expected):
    if e > 0:
        chi_square_term = (o - e)**2 / e
        chi_square_terms.append(chi_square_term)

chi_square = np.sum(chi_square_terms)

degrees_of_freedom = len(observed) - 1 - 2
reduced_chi_square = chi_square / degrees_of_freedom

p_value = 1 - chi2.cdf(chi_square, degrees_of_freedom)

print("\nChi square analysis")
print(f"Chi square = {chi_square}")
print(f"Degrees of freedom = {degrees_of_freedom}")
print(f"Reduced chi square = {reduced_chi_square}")
print(f"P-value = {p_value}")

if reduced_chi_square <= 2 and p_value > 0.05:
    print("Conclusion: The data is reasonably consistent with a normal distribution.")
else:
    print("Conclusion: The data does not strongly support a normal distribution.")

plt.hist(data, bins=bins, edgecolor="black")

x_values = np.linspace(minimum, maximum, 500)
normal_curve = norm.pdf(x_values, mean, standard_deviation)
normal_curve_scaled = normal_curve * count * bin_width

plt.plot(x_values, normal_curve_scaled)

plt.xlabel("Reaction time (s)")
plt.ylabel("Number of measurements")
plt.title(f"{name} reaction time histogram")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, chi2

data = np.array([8.16, 8.14, 8.12, 8.16, 8.18, 8.10, 8.18, 8.18, 8.18, 8.24,
8.16, 8.14, 8.17, 8.18, 8.21, 8.12, 8.12, 8.17, 8.06, 8.10,
8.12, 8.10, 8.14, 8.09, 8.16, 8.16, 8.21, 8.14, 8.16, 8.13])

name = "Taylor 12.2"

count = len(data)
mean = np.sum(data) / count

differences = []
for x in data:
    difference = x - mean
    differences.append(difference**2)

summation = np.sum(differences)
standard_deviation = (summation / (count - 1)) ** (1 / 2)

print(f"{name} mean (x̄) = {mean:.6f}")
print(f"{name} standard deviation (σ) = {standard_deviation:.6f}")

# --- 4-bin grouping with boundaries at x̄ - σ, x̄, x̄ + σ ---
# Bin 1: (-∞,  x̄ - σ)
# Bin 2: (x̄ - σ,  x̄)
# Bin 3: (x̄,  x̄ + σ)
# Bin 4: (x̄ + σ,  +∞)

b0 = -np.inf
b1 = mean - standard_deviation
b2 = mean
b3 = mean + standard_deviation
b4 = np.inf

# Expected counts from normal distribution
expected = np.array([
    norm.cdf(b1, mean, standard_deviation) - norm.cdf(b0, mean, standard_deviation),
    norm.cdf(b2, mean, standard_deviation) - norm.cdf(b1, mean, standard_deviation),
    norm.cdf(b3, mean, standard_deviation) - norm.cdf(b2, mean, standard_deviation),
    norm.cdf(b4, mean, standard_deviation) - norm.cdf(b3, mean, standard_deviation),
]) * count

bin_labels = [f"(-∞, {b1:.4f})", f"[{b1:.4f}, {b2:.4f})", f"[{b2:.4f}, {b3:.4f})", f"[{b3:.4f}, +∞)"]

print("\n4-Bin Analysis")

# Chi-square calculation using observed counts
observed = np.array([
    np.sum(data < b1),
    np.sum((data >= b1) & (data < b2)),
    np.sum((data >= b2) & (data < b3)),
    np.sum(data >= b3)
], dtype=float)

print(f"{'Bin':<30} {'Observed':>10} {'Expected':>10}")
print("-" * 52)
for label, o, e in zip(bin_labels, observed, expected):
    print(f"{label:<30} {o:>10.0f} {e:>10.4f}")

chi_square_terms = (observed - expected) ** 2 / expected
chi_square = np.sum(chi_square_terms)

# Degrees of freedom: 4 bins - 1 (normalization) - 2 (mean & sigma estimated) = 1
degrees_of_freedom = len(observed) - 1 - 2
reduced_chi_square = chi_square / degrees_of_freedom
p_value = 1 - chi2.cdf(chi_square, degrees_of_freedom)

print("\nChi-Square Analysis")
print(f"χ² = {chi_square:.4f}")
print(f"Degrees of freedom = {degrees_of_freedom}")
print(f"Reduced χ² = {reduced_chi_square:.4f}")
print(f"P-value = {p_value:.4f}")

if reduced_chi_square <= 2 and p_value > 0.05:
    print("Conclusion: The data is reasonably consistent with a normal distribution.")
else:
    print("Conclusion: The data does not strongly support a normal distribution.")

# Histogram with bin boundaries
plt.figure(figsize=(8, 5))
finite_boundaries = [b1, b2, b3]
for b in finite_boundaries:
    plt.axvline(b, color='red', linestyle='--', alpha=0.7)

x_values = np.linspace(np.min(data) - standard_deviation, np.max(data) + standard_deviation, 500)
bins_plot = [np.min(data), b1, b2, b3, np.max(data)]
plt.hist(data, bins=bins_plot, edgecolor="black", alpha=0.7, label="Observed")

bin_width_plot = standard_deviation / 2
normal_curve = norm.pdf(x_values, mean, standard_deviation)
normal_curve_scaled = normal_curve * count * bin_width_plot
plt.plot(x_values, normal_curve_scaled, label="Normal fit")

plt.xlabel("Measurement value")
plt.ylabel("Number of measurements")
plt.title(f"{name} histogram\n(red dashed lines: x̄±σ bin boundaries)")
plt.legend()
plt.show()
```
