```python
import numpy as np
x_values = np.array([20,23,23,22])
differences = []
count = len(x_values)
mean = np.sum(x_values)/count
for x in x_values:
    difference = x - mean
    differences.append(difference**2)
summation = np.sum(differences)
covariance = 1 / count * summation
print(covariance)
```

```python
import numpy as np
x_values = np.array([[20,23,23,22], [30,32,35,31]])

differences = []
count = len(x_values[0])
meanx = np.sum(x_values[0])/count
meany = np.sum(x_values[1])/count

for x, y in zip(x_values[0], x_values[1]):
    difference = (x - meanx) * (y - meany)
    differences.append(difference)
summation = np.sum(differences)
covariance = 1 / count * summation
print(covariance)
```

```python
import numpy as np
xy_values = np.array([[1,2,3,5,6,7], [5,6,6,8,8,9]])

differences = []
count = len(xy_values[0])
meanxy = np.sum(xy_values[0])/count
meanyx = np.sum(xy_values[1])/count

for x, y in zip(xy_values[0], xy_values[1]):
    difference = (x - meanxy) * (y - meanyx)
    differences.append(difference)
summationxy = np.sum(differences)
covariancexy = 1 / count * summationxy
print(f"Covariance xy = {covariancexy}")

import numpy as np
x_values = np.array([1,2,3,5,6,7])
differencesx = []
countx = len(x_values)
meanx = np.sum(x_values)/count
for x in x_values:
    differencex = x - meanx
    differencesx.append(differencex**2)
summationx = np.sum(differencesx)
covariancex = 1 / countx * summationx
print(f"Covariance x = {covariancex**(1/2)}")

import numpy as np
y_values = np.array([5,6,6,8,8,9])
differencesy = []
county = len(y_values)
meany = np.sum(y_values)/county
for y in y_values:
    differencey = y - meany
    differencesy.append(differencey**2)
summationy = np.sum(differencesy)
covariancey = 1 / county * summationy
print(f"Covariance y = {covariancey**(1/2)}")

r= covariancexy / (covariancex**(1/2) * covariancey**(1/2))
print(f"Correlation coefficient r = {r}")



```
